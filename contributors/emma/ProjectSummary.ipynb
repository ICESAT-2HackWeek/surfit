# SURFIT 

Our main goal for this project was to see if we could assess surface roughness along track-length scales comparing ATL03 and ATL06 ICESat-2 data, across two regions of interest. 

Our motivating question(s) were: 
* Are surface roughness signals hiding in ATL03 at shorter along-track length scales compared with ATL06?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Utqiagvik

```python
df_u3 = pd.read_csv('../../data/utqiagvik_ATL03_gt1l.csv')
df_u3.head()
```

```python
%matplotlib widget
plt.figure(figsize=[10, 3])
plt.scatter(df_u3['x_atc']/1000, df_u3.height, color='black',s=2)
plt.title("ATL03 Utqiagvik")
plt.xlabel('N-S along track distance (km)')
plt.ylabel('Elevation (m)')
```

## Blackrock Desert

```python
df_br3 = pd.read_csv('../../data/blackrock_ATL03_gt1l.csv') 
df_br3.head()
```

```python
plt.figure(figsize=[10, 3])
plt.scatter(df_br3['x_atc']/1000, df_br3.height, color='black',s=2)
plt.title("ATL03 Blackrock Desert")
plt.xlabel('N-S along track distance (km)')
plt.ylabel('Elevation (m)')
```
